# Health Data Visualization Assignment


**Objective:** To analyze and visualize health-related data using different plots.

 ## 1. Importing Libaraires 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python
df = pd.read_csv('Health_heart_experimental.csv')
df.head()
```

## 3. Data Cleaning and Preparation

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df = df.rename(columns={'indication': 'target', 'HR': 'thalach'})
df.head()
```

### Finding:
Unnecessary index column was removed.Column names were standardized for easier analysis. First five data set are displayed . 

## 4. Count Plot (Bar Chart)

```python
sns.countplot(x='target', data=df)
plt.title('Heart Disease Indication Count')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

### Finding:
The bar plot shows the number of patients with and without heart disease.
It helps compare class distribution in the dataset.From the plot, we can understand whether the dataset is balanced or imbalanced between healthy and heart-disease patients. It also shows more than 30000 thousand people doesn't heart disease.


## 5. Box Plot

```python
sns.boxplot(x='target', y='thalach', data=df)
plt.title('Heart Rate Distribution by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Heart Rate')
plt.show()
```

### Finding:
Heart rate distribution differs between patients with and without heart disease.



## 6. Line Plot

```python
df_sorted = df.sort_values('age')
plt.plot(df_sorted['age'], df_sorted['thalach'])
plt.title('Heart Rate Trend with Age')
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.show()
```

### Finding:
The line plot shows variation of heart rate with age.
Heart rate fluctuates but shows a general trend across age groups.Age 70 groups have lower heart rate 
Others age group have normal heart rate. 

## 7. Histogram 

```python
plt.figure()
plt.hist(df['age'], bins=10)
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.title("Histogram: Age Distribution")
plt.show()
```

### Finding:

 The histogram shows how patients are distributed across different age groups. Most patients fall within the middle-age range, indicating that heart-related checkups are more common in this group. Fewer patients are observed in the younger and older age ranges, suggesting lower hospital visits or incidence in those groups. Overall, the distribution highlights age as an important factor in the dataset and helps identify the age group with the highest patient concentration.

## Conclusion
Different visualization techniques help understand health data clearly.
 Such analysis assists in identifying patterns related to heart disease.
